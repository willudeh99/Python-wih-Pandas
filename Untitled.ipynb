```python
algo_returns_df.rename(columns= {'Date':'date', 'Algo 1':'Algo 1', 'Algo 2':'Algo 2'}, inplace=True)
```

```python
# cleaning data in sp500 dataframe
sp_500_df.isnull()
```

```python
sp_500_df = sp_500_df.dropna()
```

```python
sp_500_df
```

```python
sp_500_df.info()
```

```python
# changing the close column to a float type so data is consistent
sp_500_df['Close'] = sp_500_df['Close'].str.replace('$','')
sp_500_df['Close'] = sp_500_df['Close'].apply(pd.to_numeric)
```

```python
sp_500_df.info()
```

```python
sp_500_df['Close'] = sp_500_df['Close'].pct_change()
sp_500_df
```

```python
sp_500_df.rename(columns={'Date':'date', 'Close':'Returns'}, inplace=True)
```

```python
# Concatenate the data frames
df = pd.concat([whale_df,algo_returns_df,sp_500_df], axis='columns', join='inner')
df.head(10)
```

```python
# drop date columns
df = df.drop(columns=['date'])
```

```python
df
```

```python
# Daily returns for each portfolio
df.plot(figsize=(12,8))
```

```python
# calculating cumulative returns of data
df.head(10).sort_index()
df = df.set_index('Date')
```

```python
cumulative_df = (1 + df).cumprod()
cumulative_df.head(10)
```
